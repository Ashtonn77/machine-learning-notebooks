```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df.head()
```

```python
#groups model_year by mean
df.groupby('model_year').mean()
```

```python
#get specific column
#miles per gallon
df.groupby('model_year').mean()['mpg']
```

```python
#group by multiple columns
#grouped-by columns become indexes
df.groupby(['model_year', 'cylinders']).mean()
```

```python
df.groupby('model_year').describe()
```

```python
#working with multi-index dataframes from group-by return
year_cyl = df.groupby(['model_year', 'cylinders']).mean()
```

```python
year_cyl.loc[70]
```

```python
#getting multiple outer indexes
year_cyl.loc[[70, 80]]
```

```python
#getting one outer index and one inner index
#use tuple
year_cyl.loc[(70, 4)]
```

```python
#using cross-section
year_cyl.xs(key=70, level='model_year')
```

```python
year_cyl.xs(key=4, level='cylinders')
```

```python
#only 6 & 8 - start pruning from the dataframe level
df[df['cylinders'].isin([6, 8])].groupby(['model_year', 'cylinders']).mean()
```

```python
df[df['cylinders'].isin([6, 8])].groupby(['model_year', 'cylinders']).count()
```

```python
year_cyl.swaplevel()
```

```python
year_cyl.sort_index(level='model_year', ascending=False)
```

```python
#aggregation with multiple ...
df.agg(['std', 'mean'])
```

```python
df.agg({'mpg':['max', 'sum'], 'weight':['mean', 'std']})
```
